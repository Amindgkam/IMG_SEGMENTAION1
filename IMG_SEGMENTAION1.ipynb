```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('yara.jpg')

img_colour = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure('Original img')
plt.imshow(img_colour, cmap='gray')
plt.show()
```

```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


plt.figure('grayscale')
plt.imshow(grey, cmap='gray')
plt.show()
```

```python
ret, thresh = cv2.threshold(grey, 70, 255, cv2.THRESH_BINARY)

print(ret)

plt.figure('Binary')
plt.imshow(thresh, cmap='gray')
plt.show()
```

```python
ret, thresh2 = cv2.threshold(grey, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

print(ret)

plt.figure('OTSU')
plt.imshow(thresh2, cmap='gray')
```

```python
ret, thresh3 = cv2.threshold(grey, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_TRIANGLE)

print(ret)

plt.figure('triangle')
plt.imshow(thresh3, cmap='gray')
```

```python
img = cv2.imread('NORMAL.png')

img_colour = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure('Original img')
plt.imshow(img_colour, cmap='gray')
plt.show()
```

```python
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure('grey')
plt.imshow(grey, cmap='gray')
plt.show()
```

```python
ret, thresh = cv2.threshold(grey, 90, 255, cv2.THRESH_BINARY)

print(ret)

plt.imshow(thresh, cmap='gray')
plt.show()
```

```python
ret, thresh = cv2.threshold(grey, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_TRIANGLE)

print(ret)

plt.imshow(thresh, cmap='gray')
plt.show()
```
